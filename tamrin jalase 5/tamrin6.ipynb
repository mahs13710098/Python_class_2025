```python
lis = []
lis1=[]
lis2=[]
lis3=[]
k1 = int(input('please enter the length of your list1'))
k2 = int(input('please enter the length of your list2'))
for _ in range(k1+k2):
    j = input('please enter a string or number')
    lis.append(j)
lis1=lis[0:k1]
lis2=lis[k1:k1+k2]
print(f'first list is {lis1}')
print(f'secound list is {lis2}')
for m in range(2):
    for i in lis[m*k1:k1+(m*k2)]:
        if i not in lis[(-m+1)*k1:k1+(-m+1)*k2]:
            lis3.append(i)
print(f'different elements list is {lis3}')

```
