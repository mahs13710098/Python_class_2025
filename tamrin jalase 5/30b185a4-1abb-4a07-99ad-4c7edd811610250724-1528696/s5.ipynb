از کاربر بخواهید که 10 عدد را وارد کنید و بدون استفاده از توابع min و max مینیمم و ماکزیمم آنها را حساب کنید

```python
max_ = min_ = float(input('enter a num'))
for _ in range(9):
    num = float(input('enter a num'))
    if num > max_:
        max_ = num
    if num < min_:
        min_ = num
                
min_ , max_
```

برنامه‌ای بنویسید که اعداد ۱ تا ۲۰ را چاپ کند، اما فقط اعداد زوج را نشان دهد. از دستور continue استفاده کنید تا از چاپ اعداد فرد صرف‌نظر شود.

```python
for i in range(1,20):
    if i % 2 == 1:
        continue
    print(i)
```

برنامه‌ای بنویسید که اعداد بین ۱۰ تا 100 را چاپ کند، اما فقط اعدادی که به 7 ختم نمی‌شوند را چاپ کند.

```python
for i in range(10,100):
    if i % 10 != 7:
        print(i, end = ' * ')
```
