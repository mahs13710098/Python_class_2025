```python
lis = []
lis1=[]
lis2=[]
lis3=[]
k1 = int(input('please enter the length of your list1'))
k2 = int(input('please enter the length of your list2'))
for _ in range(k1+k2):
    j = input('please enter a string or number')
    lis.append(j)
lis1=lis[0:k1]
lis2=lis[k1:k1+k2]
for i in lis1:
    if i in lis2:
        lis3.append(i)
print(f'the first list is {lis1}')
print(f'the secound list is {lis2}')
print(f'the similarity list is {lis3}')
        
```
