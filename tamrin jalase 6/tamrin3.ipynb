```python
s = input('please enter a text: ')
q={}
i=1
m=s[0]
#for i in range(j+2,j+2+len(s[j+2:]),+1):
while i <= len(s)-1:
    j1 = i+1
    if s[i] != ' ' and s[i] != ',' and s[i] != '.':
        m = m + s[i]
        i+=1
    elif i < len(s)-1 and ((s[i] == ',' and s[j1] == ' ') or (s[i] == '.' and s[j1] == ' ')):
        q[m] = s.count(m)
        m = s[i+2]
        i+=3
    elif i != len(s)-1 and (s[i] == ' ' or s[i] == '.' or s[i] == ','):
        q[m] = s.count(m)
        m = s[i+1]
        i+=2
    if i == len(s)-1:
        q[m] = s.count(m)
        i+=1
print(s)
print(q)
```
