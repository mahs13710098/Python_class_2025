```python
BMI_Normal=23
W=float(input('please enter your weight in Kilogaram: '))
H=float(input('please enter your height in meter: '))
BMI=W/(H**2)
if BMI < BMI_Normal:
    WW = (BMI_Normal - BMI)*(H**2)
    print(f'you have {WW} kilogarams underweight ')
elif BMI == BMI_Normal:
    print('your weight is normal')
else:
    WW = (BMI - BMI_Normal)*(H**2) 
    print(f'you have {WW} kilogarams overweight ')
```
