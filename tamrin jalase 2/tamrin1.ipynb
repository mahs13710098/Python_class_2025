```python
patron = input('Patrons?')
if patron == 'Full':
    WE = float(input('Wait Estimate?'))
    if 0 <= WE <=10:
        print('yes')
    elif 10 < WE <=30:
        HU = input('Hungry?')
        if HU == 'yes':
            AL = input('Alternate?')
            if AL == 'yes':
                ra = input('Raining?')
                if ra == 'yes':
                    print('yes')
                elif ra == 'no':
                    print('no')
                else:
                    print('wrong entry') 
            elif AL == 'no':
                print('yes')
            else:
                print('wrong entry')
        elif HU == 'no':
            print('yes')
        else:
            print('wrong entry')
    elif 30 < WE <=60:
        AL = input('Alternate?')
        if AL == 'yes':
            FS = input('Fri/Sat?')
            if FS == 'yes':
                print('yes')
            elif FS == 'no':
                print('no')
            else:
                print('wrong entry')
        elif AL == 'no':
            RE = input('Reservation?')
            if RE == 'yes':
                print('yes')
            elif RE == 'no':
                BA = input('Bar?')
                if BA == 'yes':
                    print('yes')
                elif BA == 'no':
                    print('no')
                else:
                    print('wrong entry')
            else:
                print('wrong entry')
        else:
            print('wrong entry')
    elif WE > 60:
        print('no')
    else:
        print('wrong entry')
elif patron == 'some':
    print('yes')
elif patron == 'none':
    print('no')
else:
    print('wrong entry')
```
