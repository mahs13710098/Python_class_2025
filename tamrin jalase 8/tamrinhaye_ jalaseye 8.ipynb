تابعی بنویسید که ان خط اول یک فایل را بخواندد ان آرگومان ورودی تابع است

```python
#exc 2
def read_nfirst_lines(file_name,n):
    with open(file_name, 'r') as file:
        lines = file.readlines()
        return [lines[i].strip() for i in range(0,n)]
nfirst_lines = read_nfirst_lines('text1.txt',3)
nfirst_lines
```

تمرین 3: 
تابعی به نام بنویسید که دوتا فایل را بصورت ورودی دریافت کند و  تعداد کلمات مشترك آنها را برگرداند

```python
#exc 3
def findM(f_name1,f_name2):
    lst =[]
    n = 0
    r1 = open(f_name1, 'r') 
    r2 = open(f_name2, 'r') 
    import re
    list_words1 = re.split(r'[!.,\n\t?() \';:\"-]+',r1.read())
    list_words2 = re.split(r'[!.,\n\t?() \';:\"-]+',r2.read())
    unique_words1 = list(set(list_words1))
    for word in unique_words1:
        if word in list_words2:
            n += 1
            lst.append(word)
    r1.close()
    r2.close()
    return n,lst

similarity = findM('text1.txt','text2.txt')
similarity
```

تمرین 4: 
تابعی بنویسید که دو نام فایل را به عنوان ورودی دریافت کند و محتوای فایل اول را به فایل دوم کپی کند

```python
#exc 4
def copy_file(f_name1,f_name2):
    with open(f_name1,'r') as f1:
        with open (f_name2,'a') as f2:
            f2.write(f1.read())

copy_file('text1.txt','text3.txt')
```

تمرین 5 : 
تابعی بنویسید به نام  که نام یک فایل را به عنوان ورودی دریافت کند و تمام خطوط خالی آن فایل را حذف کند. سپس فایل جدید را با نام مشابه ذخیره کند

```python
#exc 5
def remove_empty_lines(f_name):
    txt =''
    with open(f_name,'r') as f:
        lines_ = f.readlines()
        for _ in range(lines_.count('\n')):
            lines_.remove('\n')
        for line in lines_:
            txt += line
        with open(f_name,'w') as f2:
            f2.write(txt)

remove_empty_lines('text4.txt')

    
```

تمرین 6 : 
تابعی بنویسید که نام یک فایل و یک کلمه را به عنوان ورودی دریافت کند و تعداد تکرار آن کلمه در فایل را برگرداند

```python
def search_word(f_name,word):
    import re
    with open(f_name,'r') as f:
        list_word = re.split(r'[!.,\n\t?() \';:\"-]+',f.read())
        word_iteration = list_word.count(word)
        return  word_iteration
    
out_put = search_word('text3.txt','and')
out_put
```

تمرین 7 : 
تابعی بنویسید که دو نام فایل را به عنوان ورودی دریافت کند و تعیین کند که آیا محتوای دو فایل دقیقاً برابر هستند یا خیر

```python
def compare_files(f_name1,f_name2):
    with open(f_name1,'r') as f1:
        with open(f_name2,'r') as f2:
            if f1.read() == f2.read():
                return 'the content of input files is the same'
            return 'the content of input files is not the same'
            
out_put1 = compare_files('text1.txt','text1.txt')
out_put2 = compare_files('text2.txt','text4.txt')
out_put1,out_put2
```

تمرین 8 : 
تابعی بنویسید که ان خط آخر یک فایل را بخواند

```python
#exc 8
def read_nlast_lines(file_name,n):
    with open(file_name, 'r') as file:
        lines = file.readlines()
        return [lines[-i].strip() for i in range(n,0,-1)]
nlast_lines = read_nlast_lines('text1.txt',2)
nlast_lines
```
