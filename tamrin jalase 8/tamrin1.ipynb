```python
from datetime import datetime
import jdatetime
with open('weather_data.txt','r') as f:
    text = f.read()
import ast
n = 'Date , AQI' + '\n'
weather_dict = ast.literal_eval(text)
for i in range(len(weather_dict['list'])):
    unix_timestamp = weather_dict['list'][i]['dt']
    dt_object = datetime.fromtimestamp(unix_timestamp)
    formatted_date = dt_object.strftime("%Y-%m-%d %H:%M:%S")
    date_object_gregorian = datetime.strptime(formatted_date, "%Y-%m-%d %H:%M:%S")
    date_object_jalali = jdatetime.datetime.fromgregorian(datetime=date_object_gregorian)
    formatted_date_jalali = date_object_jalali.strftime("%Y/%m/%d %H:%M")
    n += formatted_date_jalali+' , '+str(weather_dict['list'][i]['main']['aqi']) + '\n'
with open('weather_data_AQI.txt','w') as f2:
    f2.write(n)

```
