```python
s=0
num=input('please enter an integer number: ')
if num[0] in ["+", "-"]:
    if num[1:].isdigit():
        num = int(num)
        if num == 0:
            mean_value = 0
        else:
            for j in range(abs(num)):
                s = s + (j+1)
            mean_value = s/num
        print(f'the mean of integer numbers between zero and {num} is equal to {mean_value}')
    else:
        print('the number you entered is not integer,please try again')
elif num.isdigit():
    num = int(num)
    if num == 0:
        mean_value = 0
    else:
        for j in range(abs(num)):
            s = s + (j+1)
        mean_value = s/num
    print(f'the mean of integer numbers between zero and {num} is equal to {mean_value}')
else:
    print('the number you entered is not integer,please try again')
```
