```python
num = input('please enter an integer number: ')
sum_odd = 0
sum_five = 0
if num[0] in ["+", "-"]:
    if num[1:].isdigit():
        num = int(num)  
        if num != 0:
            for j in range(abs(num)+1):
                rem1 = j%2
                rem2 = j%5
                if rem1 == 1:
                    sum_odd = sum_odd + j
                if rem2 == 0:
                    sum_five = sum_five + j
            sum_final = sum_odd*sum_five
            print(f'the final sum value is equal to {sum_final}')    
        else:
            print('the final sum value is equal to 0') 
    else:
        print('the number you entered is not integer,please try again')
elif num.isdigit():
    num = int(num)
    if num != 0:
        for j in range(abs(num)+1):
            rem1 = j%2
            rem2 = j%5
            if rem1 == 1:
                sum_odd = sum_odd + j
            if rem2 == 0:
                sum_five = sum_five + j
        sum_final = sum_odd*sum_five
        print(f'the final sum value is equal to {sum_final}')    
    else:
        print('the final sum value is equal to 0') 
else:
    print('the number you entered is not integer,please try again')
```
